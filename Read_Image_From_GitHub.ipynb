```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

# Replace the URL with the raw image URL on GitHub
url = "https://raw.githubusercontent.com/jarif87/wordcloud-png/main/batman.png"


# Send a GET request to the URL and get the binary content
response = requests.get(url)
image_content = response.content

# Create a PIL image object from the binary data
image = Image.open(BytesIO(image_content))

# Show the image
plt.figure(figsize=(10,10))
plt.imshow(image)
plt.show()
```
